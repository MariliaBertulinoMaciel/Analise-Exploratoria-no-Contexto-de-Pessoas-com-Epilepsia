```python
#import das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#leitura do arquivo .csv
df = pd.read_csv("analise_exploratoria_pessoas_com_epilepsia.csv", sep=",", encoding='UTF-8')
```

```python
#df em série
df.head()
```

```python
#drop das colunas que não serão utilizadas
columns_drop = ['Carimbo de data/hora','autorizaccao']
df_dropcolumns = df.drop(columns_drop, axis=1)
#essas colunas que foram retiradas do df são em relação a o horário e a data que foi respondido 
# o formulário e também a coluna que mostrava que todas as pessoas autorizaram utilizar os dados para pesquisa
#como não estão relacionados diretamente ao objetivo geral da análise, foram retiradas do df

df_dropcolumns
```

```python
df_dropcolumns.info()
#nesse chamado da função info identifiquei que existem valores nulos, porém eles são irrelevantes em questão do contexto 
#inteiro da análise. Portando não vou excluí-los e nem excluir nenhuma linha, nem coluna com dados nulos.
#ainda utilizando o método info, identifiquei que precisarei converter as colunas que tem valores float em inteiro
#para fins de análise de forma mais consistente, tanto aqui no notebook, como no Tableau para 
#gerar minhas visualizações.
df2=df_dropcolumns
```

```python
#colunas que estavam em float, convertidas para inteiro com exercção da coluna # 
df2['primeira_crise_dias'] = df2['primeira_crise_dias'].round().astype('Int64')
df2['primeira_crise_meses'] = df2['primeira_crise_meses'].round().astype('Int64')
df2['ano_diagnostico'] = df2['ano_diagnostico'].round().astype('Int64')
df2['conhecimento_sobre_epilepsia'] = df2['conhecimento_sobre_epilepsia'].round().astype('Int64')
df2['dificuldade_sair_sozinhe'] = df2['dificuldade_sair_sozinhe'].round().astype('Int64')
df2['afastamento_social_filhes'] = df2['afastamento_social_filhes'].round().astype('Int64')

data_base_final =df2
```

## Respondendo os Insights através da Análise de Dados com o DataSet Final

```python

# Percentual de Pessoas do gênero feminino e masculino que participaram da pesquisa (Gráfico Pizza)


#Filtros dentro da Coluna Sentido
filtro_generom_df = data_base_final[data_base_final['genero'] == 'Masculino']
filtro_generof_df = data_base_final[data_base_final['genero'] == 'Feminino']

# Calculo do tamanho do filtro Unidirecional
total_mas = len(filtro_generom_df)
#Calculo do tamanho do filtro Bidirecional
total_fem = len(filtro_generof_df)

# Criação do Gráfico Pizza
labels = ['Masculino', 'Feminino']
sizes = [total_mas, total_fem]
colors = ['tab:purple', 'tab:blue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Porcentagem dos Gêneros')
plt.axis('equal')  

plt.show()
```
