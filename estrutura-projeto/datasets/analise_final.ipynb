```python
#drop das colunas que não serão utilizadas
columns_drop = ['Carimbo de data/hora','autorizaccao']
df_dropcolumns = df.drop(columns_drop, axis=1)
#essas colunas que foram retiradas do df são em relação a o horário e a data que foi respondido 
# o formulário e também a coluna que mostrava que todas as pessoas autorizaram utilizar os dados para pesquisa
#como não estão relacionados diretamente ao objetivo geral da análise, foram retiradas do df

df_dropcolumns
```

## Respondendo os Insights através da Análise de Dados com o DataSet Final

```python
#Agrupamento para trazer um filtro de diagnostico tardio para fins de análise

data_base_final.loc[data_base_final['primeira_crise _anos']>=18, 'diagnostico']= 'Tardio'
```

```python
# Qual a quantidade de Pessoas com Epilepsia que diagnósticaram de forma tardia com mais de 18 anos? 
filtro_diagnosticoTardio = data_base_final[data_base_final['diagnostico'] == 'Tardio']
total_tardio = len(filtro_diagnosticoTardio)
print("O total de Pessoas com diagnóstico tardio é de:", total_tardio)
```

```python
# Qual a quantidade de Pessoas com Epilepsia que fazem uso de medicamentos químicos e apresentam alergia?
data_base_final.columns = data_base_final.columns.str.strip()
filtro_alergia = data_base_final[(data_base_final['tratamento:'] == 'faz uso de remédios químicos') & (data_base_final['alergia'] == 1)]
print (f"A Quantidade de Pessoas que usam medicamentos químicos e apresentam alergia é:{len(filtro_alergia)} Pessoas")
```

```python
# Qual a quantidade de Pessoas com Epilepsia apresentam perda de memória?

filtro_memoria = data_base_final[data_base_final['perda_de_memoria']==1]
totalperdamemoria = len(filtro_memoria)
print("O total de Pessoas que apresentam perda de memória é de:", totalperdamemoria)
```

```python
# Qual a quantidade de Pessoas com Epilepsia que não se sentem confortáveis em estar ou sair para ambientes sociais sozinhes?
filtro_dificuldade_sair_so = data_base_final[data_base_final['dificuldade_sair_sozinhe']==1]
total_dificuldade_sair_so = len(filtro_dificuldade_sair_so)
print(f"O total de Pessoas que não se sentem confortáveis em sair ou estar sozinhes em ambientes públicos é de : {total_dificuldade_sair_so} Pessoas")
```

```python
#Qual a quantidade de Pessoas com Epilepsia tem sensibilidade a luz e frequentam shows?
filtro_luz = data_base_final.loc[(data_base_final['sensivel_a_luz']==1 )& (data_base_final['frequenta_shows']==1)]
sensiveisluz = len(filtro_luz)
print (f"Quantidade de Pessoas Sensíveis a Luz e que frequentam shows: {sensiveisluz} Pessoas")

#Quantidade de Pessoas Sensíveis a Som Alto e que frequentam shows
filtro_som = data_base_final.loc[(data_base_final['sensivel_a_sons']==1)& (data_base_final['frequenta_shows']==1)]
sensiveissom = len(filtro_som)
print (f"Quantidade de Pessoas Sensíveis a Som Alto e que frequentam shows: {sensiveissom} Pessoas")

#Qual a quantidade de Pessoas com Epilepsia tem sensibilidade a luz e som e frequentam shows?
filtro_luz_e_som = data_base_final.loc[(data_base_final['sensivel_a_luz']==1 ) & (data_base_final['sensivel_a_sons']==1)& (data_base_final['frequenta_shows']==1)]
sensiveis_luzesom = len(filtro_luz_e_som)
print(f"Quantidade de Pessoas Sensíveis a Luz e Som ao mesmo tempo e que frequentam shows:{sensiveis_luzesom} Pessoas")
```

```python
#Esporte mais praticado entre as pessoas com epilepsia dessa amostra
esporte_mais_comum = data_base_final['esportes'].value_counts().idxmax()

print(f"O esporte mais comum praticado é: {esporte_mais_comum}")
```

```python
# Agrupamento da coluna conhecimento_sobre epilepsia para a coluna agp_info_epilepsia
#  para definit se a opinião de acesso a informação sobre a doença é muito ou pouco

data_base_final.loc[data_base_final['conhecimento_sobre_epilepsia'] == 1,'agp_info_epilepsia'] = 'Pouco'
data_base_final.loc[data_base_final['conhecimento_sobre_epilepsia'] == 0, 'agp_info_epilepsia' ] = 'Muito'
data_base_final['agp_info_epilepsia'] .astype(str)
data_base_final
```

```python
agp_info = data_base_final["agp_info_epilepsia"].value_counts()
agp_info
```

```python
# Percentual de Pessoas que responderam sobre o acesso a informação sobre a doença ser Pouco e Muito

#Filtros dentro da Coluna agp_info_epilepsia
filtro_infoP = data_base_final[data_base_final['agp_info_epilepsia'] == 'Pouco']
filtro_infoM = data_base_final[data_base_final['agp_info_epilepsia'] == 'Muito']

# Calculo do tamanho do filtro para pouca informação
total_Pouco = len(filtro_infoP)
#Calculo do tamanho do filtro para muita informação
total_Muito = len(filtro_infoM)

# Estrutura do Gráfico
labels = ['Pouca Informação', 'Muita Informação']
sizes = [total_Pouco, total_Muito]
colors = ['tab:green', 'tab:purple']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Percentual de Opinião em relação ao Acesso à informação sobre Epilepsia ')
plt.axis('equal')  

plt.show()
```

```python


# Filtro na coluna alguem_na_genetica
filtro_genetica = data_base_final['alguem_na_genetica'].value_counts()

# Estrutura do Gráfico de Barras (Horizontal)
filtro_genetica.plot(kind='barh', color=['purple', 'green'])

plt.xlabel('Quantidade de Pessoas com Alguém na genética')
plt.ylabel('Parentesco ou não')
plt.title('Distribuição de Alguém na Genética')
plt.show()

```

```python
# Quantas pessoas receberam diagnóstico depois dos anos 2000?
filtro_diagnostico_apos_anos2000 = data_base_final.loc[data_base_final['ano_diagnostico']>=2000]
total_anos2000= len(filtro_diagnostico_apos_anos2000)
print (f" A quantidade de pessoas que receberam diagnóstico do ano 2000 em diante foi: {total_anos2000} Pessoas")
```

```python
# Percentual de Pessoas que sentiram o afastamento social após diagnóstico? 

#Filtros para determinar o valor de totais de pessoas que responderam sim e não.
filtro_afastamento_social_sim = data_base_final[data_base_final['afastamento_social']==1]
filtro_afastamento_social_nao = data_base_final[data_base_final['afastamento_social']==0]

#Calculando o tamanho do filtro
total_sim = len(filtro_afastamento_social_sim)
total_nao= len(filtro_afastamento_social_nao)

# Estrutura do Gráfico
labels = ['Sim','Não']
sizes = [total_sim, total_nao]
colors = ['tab:green', 'tab:purple']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Percentual de Pessoas em relação ao Afastamento Social Após o diagnóstico')
plt.axis('equal')  

plt.show()

```

```python
# dentro dessa afastamento social quantas pessoas são do gênero feminino e quantas do gênero masculino
filtro_afastamentosocial_fem = data_base_final[(data_base_final['genero'] == 'Feminino') & (data_base_final['afastamento_social'] == 1)]
total_feminino= len(filtro_afastamentosocial_fem)
print(f"O total de Mulheres que sentiram afastamento social depois do diagnóticos é de: {total_feminino} Mulheres")

filtro_afastamentosocial_mas = data_base_final[(data_base_final['genero'] == 'Masculino') & (data_base_final['afastamento_social'] == 1)]
total_masculino= len(filtro_afastamentosocial_fem)
print(f"O total de Homens que sentiram afastamento social depois do diagnóticos é de: {total_masculino} Homens")
```

```python

# Percentual de Pessoas do gênero feminino e masculino que participaram da pesquisa (Gráfico Pizza)


#Filtros dentro da Coluna Gênero
filtro_generom_df = data_base_final[data_base_final['genero'] == 'Masculino']
filtro_generof_df = data_base_final[data_base_final['genero'] == 'Feminino']

# Calculo do tamanho do filtro Masculino
total_mas = len(filtro_generom_df)
#Calculo do tamanho do filtro Feminino
total_fem = len(filtro_generof_df)

# Estrutura do Gráfico
labels = ['Masculino', 'Feminino']
sizes = [total_mas, total_fem]
colors = ['tab:green', 'tab:purple']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Porcentagem dos Gêneros')
plt.axis('equal')  
plt.show()
```

```python
# Convertendo a coluna ano_diagnostico para datatime (estava em float), trazendo apenas o parâmetro ano.

data_base_final['ano_diagnostico'] = pd.to_datetime(data_base_final['ano_diagnostico'], format='%Y')

# Filtrando cada ano existente na coluna ano_diagnostico
data_base_final['Anos'] = data_base_final['ano_diagnostico'].dt.year

data_base_final['Anos'] = data_base_final['Anos'].round().astype('Int64')
data_base_final['ano_diagnostico'] = pd.to_datetime(data_base_final['ano_diagnostico'], format='%Y')
# Variável que guarda um agrupamento por ano faz a contagem de cada ano e realiza um sort do index para exibir 
# em ordem crescente
diagnostico_por_ano = data_base_final['Anos'].value_counts().sort_index()

#Estrutura do Gráfico no Seaborn
sns.set(style="whitegrid")
 
plt.figure(figsize=(10, 6))  
sns.lineplot(x=diagnostico_por_ano.index, y=diagnostico_por_ano.values, marker='o')
plt.xlabel('Ano de Diagnóstico')
plt.ylabel('Número de Diagnósticos')
plt.title('Número de Diagnósticos por Ano')
plt.show()
```

```python
# Identificando a quantidade de tipos de crises existentens e filtrando apenas a quantidade de pessoas que tem crise de ausência
tipos_de_crise = data_base_final [' tipos_de_crises '].value_counts()
print(f"A quantidade de tipos de crises existentes nessa amostra é de: {len(tipos_de_crise)} Tipos de Crises")


filtro_c_ausencia = data_base_final[(data_base_final[' tipos_de_crises '] == 'Crise generalizada de Ausência')]
print(f"A quantidade de pessoas que apresentam crise de ausência é de : {len(filtro_c_ausencia)} ")
```

```python
#Filtrando os 3 primeiros tipos de diagnosticos neurodivergente que são apresentados pelas pessoas da amostra
top3_outros_diagnosticos_neurodivergente = data_base_final.sort_values(by='diagnostico_neurodivergente', ascending=False).head(3)

# Estrutura do Gráfico
plt.figure(figsize=(10, 6))
plt.bar(top3_outros_diagnosticos_neurodivergente['tipos_de_crises'], top3_outros_diagnosticos_neurodivergente['diagnostico_neurodivergente'], color='purple')
plt.xlabel('Tipos de Crises Epilépticas')
plt.ylabel('Outros diagnósticos neurodivergentes')
plt.title('Top 3 diagnósticos divergentes associados a epilepsia')
plt.xticks(rotation=45, ha='right')

# Laço for iterando os tipos de diagnósticos e trazendo os tipos de crises que são mais associados a esse tipo de diagnostico neurodivergente
for  i, diagnostico in enumerate(top3_outros_diagnosticos_neurodivergente['tipos_de_crises']):
    rota = top3_outros_diagnosticos_neurodivergente.iloc[i]['tratamento:']
    plt.annotate(rota, (diagnostico, top3_outros_diagnosticos_neurodivergente.iloc[i]['tipos_de_crises']), ha='center', va='bottom')

plt.show()
```

```python
#Agrupamento da Coluna idade por grupos de faixas etárias, exibindo uma nova coluna chamada grupo_idade. 

data_base_final.loc[data_base_final['idade'] < 10, 'grupo_idade'] = '0-9'
data_base_final.loc[(data_base_final['idade'] >= 10) & (data_base_final['idade'] < 16), 'grupo_idade'] = '10-15'
data_base_final.loc[(data_base_final['idade'] >= 16) & (data_base_final['idade'] < 19), 'grupo_idade'] = '16-18'
data_base_final.loc[(data_base_final['idade'] >= 19) & (data_base_final['idade'] < 26), 'grupo_idade'] = '19-25'
data_base_final.loc[(data_base_final['idade'] >= 26) & (data_base_final['idade'] < 40), 'grupo_idade'] = '26-39'
data_base_final.loc[(data_base_final['idade'] >= 40) & (data_base_final['idade'] < 61), 'grupo_idade'] = '40-60'
data_base_final.loc[(data_base_final['idade'] >= 51 ) & (data_base_final['idade'] < 60), 'grupo_idade'] = '45-59'
data_base_final.loc[data_base_final['idade'] >= 60, 'grupo_idade'] = '60+'
data_base_final
```

```python
#Quantidade de Pessoas por Grupo de Idade
grupo_idade = data_base_final["grupo_idade"].value_counts()
grupo_idade
```

```python
#Estrutura do Gráfico que exibe a quantidade de pessoas por grupo de faixa etária
plt.figure(figsize=(8, 6))

cor_desejada = 'purple'

#Instanciando os grupos de faixa etária com a função countplot do Seaborn
sns.countplot(data=data_base_final,
              x='grupo_idade',
              color=cor_desejada)
plt.title('Quantidade de Pessoas com Epilepsia por Grupos de Faixa Etária')
plt.xlabel('Grupos de Faixas etárias')
plt.ylabel('Quantidade de Pessoas')
plt.show()
```
